1. Dalam suatu liga sepak bola, dipilih secara acak 10 pemain dari klub yang berbeda dan diketahui jumlah gol yang mereka raih dalam musim ini adalah: 19, 10, 10, 12, 13, 15, 14, 13, 9, 17. Lakukan analisis terhadap sebaran data dari dataset yang diberikan. Manakah ukuran sebaran yang paling mampu menggambarkan sebaran data dengan baik?

a. Range

```python
def calculate_range(data):
  nilai_tertinggi = max(data)
  print(f"Nilai tertinggi: {nilai_tertinggi}")

  nilai_terendah = min(data)
  print(f"Nilai terendah: {nilai_terendah}")

  # Selisih antara nilai tertinggi dan terendah
  data_range = max(data) - min(data)

  print(f"\n")
  print(f"Range: {nilai_tertinggi} - {nilai_terendah} = {data_range}")



# Inisialisasi dataset
data = [19, 10, 10, 12, 13, 15, 14, 13, 9, 17]
print(f"Data: {data}\n")

import matplotlib.pyplot as plt

# Create a histogram
plt.hist(data, bins=30, edgecolor='black', alpha=0.5, color='green')

# Define the horizontal line coordinates
y = 1.5 # The y-coordinate where the line will be drawn
x1 = 4/70*1  # The x-coordinate of the start point
x2 = 66/70*1  # The x-coordinate of the end point

# Add a horizontal line
plt.axhline(y=y, xmin=x1, xmax=x2, color='red', linestyle='dashed', linewidth=1, label=f'Range = {calculate_range(data)}')

# Add vertical line
plt.axvline(x=min(data), color='blue', linestyle='dashed', linewidth=1, label=f'Min Data: {min(data):.2f}')
plt.axvline(x=max(data), color='blue', linestyle='dashed', linewidth=1, label=f'Max Data: {max(data):.2f}')

# Add labels and title
plt.xlabel('Jumlah Gol')
plt.ylabel('Frequency')
plt.title('Range')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

b. Variansi

```python
# Fungsi untuk menghitung rata-rata
def hitung_rata_rata(data):
  return round(sum(data) / len(data), 2)

# Fungsi untuk menghitung variansi
def hitung_variansi(data):
  # Rata-rata
  rata_rata = hitung_rata_rata(data)
  print(f"Rata-rata: {rata_rata}")

  # Kuadrat selisih antara setiap nilai data dengan rata-rata
  print(f"Kuadrat selisih setiap nilai data dengan rata-rata:")
  kuadrat_selisih = []
  for x in data:
    selisih = round(x - rata_rata, 2)
    kuadrat_selisih_per_x = round(selisih**2, 2)
    kuadrat_selisih.append(kuadrat_selisih_per_x)
    print(f"({x} - {rata_rata})**2 = {selisih}**2 = {kuadrat_selisih_per_x}")

  print(f"Total kuadrat selisih: {round(sum(kuadrat_selisih), 2)}")
  print(f"Jumlah total data: {len(data)}")

  # Jumlah kuadrat selisih antara setiap nilai data dengan rata-rata, kemudian dibagi dengan jumlah total data
  variansi = round(sum(kuadrat_selisih) / len(data), 2)

  print(f"\n")
  print(f"Variansi: {round(sum(kuadrat_selisih), 2)} / {len(data)} = {variansi}")

  return variansi

print(f"Data: {data}")
variansi = hitung_variansi(data)
```

c. Standar Deviasi

```python
# Fungsi untuk menghitung standar deviasi
def hitung_deviasi_standar(data):
  variansi = hitung_variansi(data)
  deviasi_standar = round(variansi**0.5, 2)
  print(f"\n")
  print(f"Deviasi Standar: {variansi}**0.5 = {deviasi_standar}")

  return deviasi_standar

# Inisialisasi dataset
print(f"Data: {data}")

deviasi_standar = hitung_deviasi_standar(data)
rata_rata = hitung_rata_rata(data)

import matplotlib.pyplot as plt
import numpy as np

# Buat histogram
plt.hist(data, bins=30, edgecolor='black', alpha=0.5, color='green')

# Tambahkan garis vertikal pada deviasi standar
plt.axvline(x=rata_rata, color='red', linestyle='dashed', linewidth=1, label=f'Rata-rata = {rata_rata}')
plt.axvline(x=rata_rata - deviasi_standar, color='blue', linestyle='dashed', linewidth=1, label=f'{rata_rata-deviasi_standar:.2f}')
plt.axvline(x=rata_rata + deviasi_standar, color='blue', linestyle='dashed', linewidth=1, label=f'{rata_rata+deviasi_standar:.2f}')

# Tambahkan label dan judul
plt.xlabel('Jumlah Gol')
plt.ylabel('Frekuensi')
plt.title(f'Deviasi Standar = {deviasi_standar:.2f}')

# Tambahkan legenda
plt.legend()

# Tampilkan plot
plt.show()
```

Berdasarkan percobaan di atas, dapat disimpulkan bahwa variansi dan deviasi standar merupakan metode yang paling efektif untuk menggambarkan sebaran data secara komprehensif. Kedua metode ini memperhitungkan seluruh distribusi nilai, bukan hanya dua nilai ekstrem, sehingga lebih mampu mengidentifikasi outlier atau nilai yang signifikan jauh dari mayoritas data.

2. Dalam tim olahraga, tinggi pemain adalah: 180 cm, 185 cm, 175 cm, 190 cm, 170 cm, 195 cm, 165 cm. Bagaimana sebaran data dari tinggi pemain ini?

a. Range

```python
import matplotlib.pyplot as plt
import numpy as np

# Fungsi untuk menghitung rentang
def hitung_rentang(data):
  nilai_tertinggi = max(data)
  nilai_terendah = min(data)

  # Selisih antara nilai tertinggi dan terendah
  rentang_data = nilai_tertinggi - nilai_terendah

  return rentang_data

# Inisialisasi dataset
tinggi = [180, 185, 175, 190, 170, 195, 165]

# Hitung rentang
rentang = hitung_rentang(tinggi)

# Buat histogram dengan warna gradasi biru
plt.hist(tinggi, bins=30, edgecolor='black', alpha=0.7, color='skyblue')

# Tambahkan garis vertikal pada rentang
plt.axvline(x=min(tinggi), color='blue', linestyle='dashed', linewidth=2, label=f'Data Min: {min(tinggi):.2f}')
plt.axvline(x=max(tinggi), color='red', linestyle='dashed', linewidth=2, label=f'Data Max: {max(tinggi):.2f}')

# Tambahkan garis horizontal untuk rentang
y = 1.5 # The y-coordinate where the line will be drawn
x1 = min(tinggi) / max(tinggi)  # The x-coordinate of the start point
x2 = 1  # The x-coordinate of the end point
plt.plot([min(tinggi), max(tinggi)], [y, y], color='purple', linestyle='dashed', linewidth=2, label=f'Rentang = {rentang:.2f}')

# Tambahkan label dan judul
plt.xlabel('Tinggi Pemain', fontsize=12)
plt.ylabel('Frekuensi', fontsize=12)
plt.title(f'Rentang = {rentang:.2f}', fontsize=14)

# Tambahkan legenda
plt.legend()

# Tampilkan plot
plt.show()
```

b. Variansi

```python
# Fungsi untuk menghitung rata-rata
def calculate_mean(data):
  return sum(data) / len(data)

# Fungsi untuk menghitung variansi
def calculate_variance(data):
  # Rata-rata
  mean = calculate_mean(data)
  print(f"Rata-rata: {mean}")

  # Kuadrat selisih antara setiap nilai data dengan rata-rata
  print(f"Kuadrat selisih setiap nilai data dengan rata-rata:")
  squared_diff = []
  for x in data:
    squared_diff_per_x = (x - mean)**2
    squared_diff.append( squared_diff_per_x )
    print(f"({x} - {mean})**2 = {(x - mean)}**2 = {squared_diff_per_x}")

  print(f"Total kuadrat selisih: {sum(squared_diff)}")
  print(f"Jumlah total data: {len(data)}")

  # Jumlah kuadrat selisih antara setiap nilai data dengan rata-rata, kemudian dibagi dengan jumlah total data
  variance = sum(squared_diff) / len(data)

  print(f"\n")
  print(f"Variansi: {sum(squared_diff)} / {len(data)} = {variance}")

  return variance


# Inisialisasi dataset
print(f"Data: {tinggi}")

variance_tinggi = calculate_variance(tinggi)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Fungsi untuk menghitung standar deviasi
def hitung_deviasi_standar(data):
  variansi = hitung_variansi(data)
  deviasi_standar = round(variansi**0.5, 2)
  print(f"\n")
  print(f"Deviasi Standar: {variansi}**0.5 = {deviasi_standar}")

  return deviasi_standar

# Inisialisasi dataset
print(f"Data: {tinggi}")

deviasi_standar = hitung_deviasi_standar(tinggi)
rata_rata = hitung_rata_rata(tinggi)

# Buat histogram dengan warna gradasi biru
plt.hist(tinggi, bins=30, edgecolor='black', alpha=0.7, color='skyblue')

# Tambahkan garis vertikal pada deviasi standar
plt.axvline(x=rata_rata, color='red', linestyle='dashed', linewidth=2, label=f'Rata-rata = {rata_rata}')
plt.axvline(x=rata_rata - deviasi_standar, color='blue', linestyle='dashed', linewidth=2, label=f'{rata_rata-deviasi_standar:.2f}')
plt.axvline(x=rata_rata + deviasi_standar, color='blue', linestyle='dashed', linewidth=2, label=f'{rata_rata+deviasi_standar:.2f}')

# Tambahkan label dan judul
plt.xlabel('Tinggi Pemain', fontsize=12)
plt.ylabel('Frekuensi', fontsize=12)
plt.title(f'Deviasi Standar = {deviasi_standar:.2f}', fontsize=14)

# Tambahkan legenda
plt.legend()

# Tampilkan plot
plt.show()
```
